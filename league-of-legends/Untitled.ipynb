### 3. Preprocess

```python
# train test split
from sklearn.model_selection import train_test_split
#bo cac cot nay di 
drop_columns = ['gameId','creationTime','gameDuration','seasonId','winner']
X = data.drop(columns=drop_columns) #future x la tong x train va x test tuong tu y  
y = data['winner'] #du doan 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
sns.heatmap(data.corr())
```

### 4. Create Model

```python
# Classification
# Logistic Regressio
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)#tao model voi x trai y train

# Random Forest

# SVM
```

```python
lr_model.predict(X_test)
```

```python
prediction_test = lr_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confu = confusion_matrix(y_test,prediction_test)
```

```python
sns.heatmap(confu, annot=True, fmt='g')
plt.xlabel('Predict')
plt.ylabel('Actual')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, prediction_test))
```
